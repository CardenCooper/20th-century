```python
from textblob import TextBlob
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import nltk
import nltk
import re
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.corpus import stopwords
from collections import Counter
sns.set()
```

```python
# Import txt file

with open('Key_Events_Of_The_20th_Century_CLEAN.txt', 'r', errors='ignore') as file: 
   data = file.read().replace( '\n', ' ')
```

```python
# Sentence tokenization

from nltk.tokenize import sent_tokenize
tokenized_sent = sent_tokenize(data)
print(tokenized_sent) 
```

```python
# Word tokenization

from nltk.tokenize import word_tokenize
tokenized_word = word_tokenize(data)
print(tokenized_word)
```

```python
# finding the frequency distribution

from nltk.probability import FreqDist
dist_words = FreqDist(tokenized_word)
print(dist_words) 
```

```python
dist_words.most_common(10)
```

```python
# Frequency Distribution Plot

plt.figure(figsize=(8, 3))
dist_words.plot(10, cumulative=False)
plt.show()
```

```python
stop_words = set(stopwords.words('english'))
```

```python
# Removing stopwords in words
filtered_words = []  # creates an empty list

for word in tokenized_word:
    if word not in stop_words:
        filtered_words.append(word)
```

```python
# Create a new FreqDist for filtered–words

dist_words_filter = FreqDist(filtered_words)
print (dist_words_filter)
```

```python
# plotting the new filtered words

plt.figure(figsize=(8, 3))
dist_words_filter.plot(10, cumulative=False)
plt.show()
```

```python
# Substitute all punctuation marks with a space

sans_punc = re.sub("[^a-zA-Z]",  # Search for all non-letters
                        " ",        # Replace all non-letters with spaces
                        str(filtered_words))
```

```python
# Word tokenization

tokenized_word_2 = word_tokenize(sans_punc)
print (tokenized_word_2)
```

```python
# Create a new FreqDist

dist_words_filter_2 = FreqDist(tokenized_word_2) 
```

```python
# plotting the new word filter

plt.figure(figsize=(8, 3))
dist_words_filter_2.plot(10, cumulative=False)
plt.show()
```

Now instead of having only 2 real words we have 8 and you can see a lot of WW2 related words in the filter

```python
# creating tag list

blob = TextBlob(data)
tags = blob.tags   # list of (word, pos)

df_tags = pd.DataFrame(tags, columns=["word","pos"])

# Filter out tokens that are purely punctuation (optional but helpful)
df_tags = df_tags[df_tags["word"].str.match(r"[A-Za-z0-9'-]+")]
df_tags["word_lower"] = df_tags["word"].str.lower()
df_tags.head()
```

```python
# creating plot of top 10 POS tags in article

pos_counts = df_tags["pos"].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=pos_counts.values, y=pos_counts.index, orient="h")
plt.xlabel("Count")
plt.ylabel("POS tag")
plt.title("Top 10 POS tags in article")
plt.show()
```

```python
# defining POS tag groups

noun_tags = ['NN','NNS','NNP','NNPS']
verb_tags = ['VB','VBD','VBG','VBN','VBP','VBZ']
adj_tags  = ['JJ','JJR','JJS']

def top_words_for(tag_list, topn=15):
    subset = df_tags[df_tags["pos"].isin(tag_list)]
    top = subset["word_lower"].value_counts().head(topn)
    return top
```

```python
# Nouns
nouns_top = top_words_for(noun_tags, 15)
plt.figure(figsize=(10,6))
sns.barplot(x=nouns_top.values, y=nouns_top.index)
plt.title("Top 15 Nouns")
plt.xlabel("Count")
plt.ylabel("Noun")
plt.show()
```

```python
# Verbs
verbs_top = top_words_for(verb_tags, 15)
plt.figure(figsize=(10,6))
sns.barplot(x=verbs_top.values, y=verbs_top.index)
plt.title("Top 15 Verbs")
plt.xlabel("Count")
plt.ylabel("Verb")
plt.show()
```

```python
# Adjectives
adjs_top = top_words_for(adj_tags, 15)
plt.figure(figsize=(10,6))
sns.barplot(x=adjs_top.values, y=adjs_top.index)
plt.title("Top 15 Adjectives")
plt.xlabel("Count")
plt.ylabel("Adjective")
plt.show()
```

```python
countries_df = pd.read_csv("countries_list_20th_century_1.5.csv")
```

```python
# Inspecting the file to see column names

print(countries_df.head())
print(countries_df.columns)
```

```python
country_list = countries_df["country_name"].str.strip().unique().tolist()
print(country_list[:20]) 
```

```python
# Syntax to make sure that countries like "United States" are counted correctly with the space in the middle

country_counts = {}
for country in countries_df["country_name"].str.strip().unique():
    pattern = r'\b' + re.escape(country.strip()) + r'\b'
    matches = re.findall(pattern, listToStr, flags=re.IGNORECASE)
    count = len(matches)
    if count > 0:
        country_counts[country] = count

df_countries = pd.DataFrame(country_counts.items(), columns=["Country", "Times mentioned"])
df_countries = df_countries.sort_values(by="Times mentioned", ascending=False)
print(df_countries.head(20))
```

```python
# plotting top 20 most commonly mentioned countries 

plt.figure(figsize=(10,6))
sns.barplot(data=df_countries.head(20), x="Times mentioned", y="Country", palette="viridis")
plt.title("Top 20 Countries Mentioned in the Text")
plt.show()
```

## Findings are about as expected the top 5 countries are all countries who were heavily involved in WW2 which was the biggest event of the 20th century.
